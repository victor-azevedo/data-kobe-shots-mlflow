```python
def conform_data(df: pd.DataFrame) -> pd.DataFrame:
    df_conformed = df.dropna()
    df_conformed = df_conformed.loc[df['shot_type'] == "2PT Field Goal"]
    df_conformed = df_conformed[['lat','lon', 'minutes_remaining', 'period', 'playoffs','shot_distance', 'shot_made_flag' ]]
    return df_conformed
```

```python
if not os.path.isfile("../../Data/processed/data_filtered"):
    df_2pt = conform_data(df)
    df_2pt.to_parquet(path="../../Data/processed/data_filtered")
    print("data_filtered.parquet created")
else:
    df_2pt = pd.read_parquet(path="../../Data/processed/data_filtered")
    print("data_filtered.parquet loaded")

df_2pt.sample(10)
```

```python
from sklearn.preprocessing import StandardScaler

from sklearn.model_selection import train_test_split
from sklearn.utils.class_weight import compute_class_weight
from sklearn.model_selection import cross_val_score 
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC

from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score

import numpy as np
```

```python
SEED = 5
np.random.seed(SEED)
```

```python
data = df_2pt.copy()
x = df_2pt.drop('shot_made_flag', axis=1)
y = df_2pt['shot_made_flag']


x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, stratify = y)
```

```python
print(y.value_counts(normalize=True))

weight = compute_class_weight(class_weight="balanced",classes=np.unique(y) , y=y)
class_weight = {0: weight[0], 1: weight[1]}
class_weight
```

```python
scaler = StandardScaler()
x_scaler = scaler.fit_transform(x)

# Scale the features using StandardScaler
x_train_scaler = scaler.fit_transform(x_train)
x_test_scaler = scaler.transform(x_test)
```

### Baseline

```python
import numpy as np
y_baseline = np.ones(y_test.shape[0])

baseline_report = classification_report(y_test, y_baseline, zero_division=0)
print(baseline_report)
```

### Regressao logistica

```python
params = {
        'penalty':  ['l1','l2', 'elasticnet'],
        'C': [0.01, 0.1, 0.5, 1, 2, 10],
    }

cv = StratifiedKFold(n_splits=10)

model_template = LogisticRegression(solver='saga',n_jobs=-1, max_iter=10000, class_weight=class_weight)

clf = GridSearchCV(
        model_template,
        params,
        cv=cv,
        scoring=['f1', 'precision', 'recall'],
        refit='f1',
        return_train_score=True,
        n_jobs=-1
    )

clf.fit(x_train_scaler, y_train)
best_lr_model = clf.best_estimator_

best_lr_model.fit(x_train_scaler,y_train)

y_pred = best_lr_model.predict(x_test_scaler)
lr_report = classification_report(y_test, y_pred)

print(lr_report)
print(best_lr_model)
```

### SVC

```python
svc = SVC(class_weight=class_weight)

svc.fit(x_train_scaler, y_train)

y_pred = svc.predict(x_test_scaler)
svc_report = classification_report(y_test, y_pred)

print(svc_report)
print(svc)
```
